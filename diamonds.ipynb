```python
mean_hp = df.groupby("cut")["price"].mean()
mean_hp.plot(kind = "bar", figsize = (6,3));
```

```python
mean_hp = df.groupby("clarity")["price"].mean()
mean_hp.plot(kind = "bar", figsize = (6,3));
```
